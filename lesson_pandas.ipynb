### Pandas Datastructures
* Series - single dimensional datastructure. Similare to a list or a single column
* homogeneous (of the same data type)
* DataFrame - multidimensional table
* heterogeneous (of many datatypes)

```python
# Creating a series with a list literal
ser1 = pd.Series([10, 20, 30, 40], index=["A", "B", "C", "D"])

print(ser1)
```

```python
# Creating a series from a preexisting dictionary
d = {
    "b": 1,
    "a": 0,
    "c": 2,
    "d":100
}

ser2 = pd.Series(d)
```

```python
d1 = {
    "a":0.0,
    "b":1.0,
    "c":2.0
}

ser3 = pd.Series(d1)

print(ser3)
```

```python
calories = {
    'day1': 700,
    'day2':2500,
    'day3': 1200
}

# use indexes to chose certain values for a series
ser4 = pd.Series(calories, index=['day1', 'day3'])

print(ser4)
```

```python
# If provided a scalar values, value will repreat to fill al index labels
ser5 = pd.Series(5.0, index=["a", "b", "c"])

print(ser5)
```

### Frequency Analysis - Value counts

```python
# Sample DataFrame
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Books', 'Clothing']}

cat_df = pd.DataFrame(data)

# .value_counts
cat_df.value_counts()
```

```python
# Create the dataframe
sales_data={"Devices":['Laptop','iPhone','LED','LCD','Smart-Phone','Washing-Machine'],
           'Brand':['Lenovo','Apple','Samsung','Samsung','Samsung','Whirpool'],
           'Sales':[1000,2000,4000,2000,1000,4000],
           'Profit':[500,1000,1000,1500,1000,1500],
           'Pices left':[5000,4000,4000,5000,5000,1000]}

sales_df = pd.DataFrame(sales_data)

# Get frequency count for Brands
Frequency = sales_df['Brand'].value_counts()
# print(Frequency)

frequent_product = Frequency[Frequency > 1].index[0]

print(Frequency)
print(f'This item appears more than once: {frequent_product}')
```

### Sort using Pandas
* .sort_values()

```python
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
       'Score 1' : [56,86,77,45],
       'Score 2' : [50,96,60,30]}

grades_df = pd.DataFrame(data)

grades_df.head()
```

```python
grades_df = grades_df.assign(addess=['NYC', 'NJ', 'CA', 'PA'])

grades_df.head()
# sort_values(). sorts by ascending by default
print(grades_df.sort_values(by='Score 1'))
print(grades_df.sort_values(by='Score 1', ascending=False))
cars = pd.read_json('data/cars.json')

cars.head(15)
print(cars.sort_values(by='quantity', ascending=True))
print(cars.sort_values(by=['quantity', 'Car'], ascending=[False, True]))
```

### Pandas Date & Time
* 'pd.to_datetime() - convert different data types onto datetime object
* 'pd.date_range() - used different a range and time
* 'pd.timestamp() - generate the time expressed as number of seconds since 01/01/1970

```python
# Generate a range of dates from start to end, default frequency
# periods = number of dates you want between start and end (inclusive)
date_range = pd.date_range(start='04/05/2026', end='05/05/2026' , periods=7)

print(date_range)
```

```python
start_date = '01/01/2023'
num_days=30

date_range = pd.date_range(start=start_date, periods=num_days, freq='D')

np.random.seed(42) # seed means we are all using the SAME random algorithm
stock_prices = np.random.normal(loc=100, scale=5, size=num_days).round(2)

stock_df = pd.DataFrame({'Date': date_range, 'StockPrice':stock_prices})

stock_df
```

```python
url='https://raw.githubusercontent.com/bprasad26/lwd/master/data/tesla_stock_prices.csv'
# using parse_dates to change type from string into date
tesla_df = pd.read_csv('data/tesla_stocks.csv')

tesla_df.head()
# tesla_df.info()
```

```python
# using parse_dates to change type from string into date
tesla_df = pd.read_csv('data/tesla_stocks.csv', parse_dates=['Date'])

tesla_df.head()
```

```python
# Add new columns to dataframe based off of dates
tesla_df['Day'] = tesla_df['Date'].dt.day
tesla_df['Month'] = tesla_df['Date'].dt.month
tesla_df['Year'] = tesla_df['Date'].dt.year
tesla_df['day_name'] = tesla_df['Date'].dt.day_name()

tesla_df.head(10)
```

```python
# Grab max or min date from a dataset
least_recent = tesla_df['Date'].min()
most_recent = tesla_df['Date'].max()

print(most_recent)
print(least_recent)
```

```python
# Group by month and year and sum the volume
monthly_sales = tesla_df.groupby(['Year', 'Month'])['Volume'].sum().reset_index()

monthly_sales.head(15)
```

### Filter using loc[ ] or query()

```python
# filtered_df_date = telsa_df.loc[tesla_df['Date'] >= '2010-08-01']
filtered_df_date = tesla_df.loc[(tesla_df['Date'] >= '2010-08-01') & (tesla_df['Date'] <= '2010-12-31')]
# created a df of a range of dates using conditionals
# conditional filtering
filtered_df_date
```

```python
filtered_df = tesla_df.query("Date >= '2010-08-01' and Date <= '2010-12-31'")

filtered_df
```
