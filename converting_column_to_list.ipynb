## Converting Columns to List

```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_json('data/cars.json')
```

```python
df.head()
```

### Converting Pandas Column to list using .tolist()

```python
col_list = df['Car'].values.tolist()
print('CarList:', col_list)
```

```python
df.Car.values.tolist()
```

### Converting Pandas column to a Numpy array

```python
col_numpy_array = df['Car'].to_numpy()
print(type(col_numpy_array))
```

```python
df[df.columns[0]].values.tolist()
```
